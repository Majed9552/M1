```python
# 0.  Imports & notebook beautification
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# prettier plots
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
%config InlineBackend.figure_format = 'retina'
```

```python
# 1.  Load Titanic-Dataset.csv
# ==========================================
df = pd.read_csv("Titanic-Dataset.csv")   # be sure file is in the same folder
print("Shape:", df.shape)
df.head()
```

```python
# 2.  First look at the data
# ==========================================
display("Head", df.head())
display("Tail", df.tail())
display("Data types", df.dtypes.to_frame("dtype"))
display("Numerical summary", df.describe())
```

```python
# 3.  Clean & enrich the data
# ==========================================
# 3-a  categorical Pclass
df["PclassCategory"] = df["Pclass"].astype("category").cat.rename_categories(
    {1: "Upper", 2: "Middle", 3: "Lower"}
)

# 3-b  categorical Cabin (first letter, or 'Unknown')
df["CabinCategory"] = df["Cabin"].fillna("Unknown").str[0]

# 3-c  missing ages  median (robust to outliers)
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)

# 3-d  family size
df["Family"] = df["SibSp"] + df["Parch"] + 1   # +1 for the passenger himself

df.head()
```

```python
# 4.  Extreme cases
# ==========================================
# highest fare
highest_fare_passengers = df.loc[df["Fare"].idxmax()].to_frame("max_fare_passenger")

# largest family
largest_family_passengers = df[df["Family"] == df["Family"].max()]
print("Highest fare passenger:\n", highest_fare_passengers.T)
print("\nLargest family passengers:\n", largest_family_passengers[["Name", "Family"]])
```

```python
# 5.  Series & DataFrame gymnastics
# ==========================================
# 5-a  Series of names
names_series = pd.Series(df["Name"].values, index=df.index)

# 5-b  mini DataFrame
names_ages_df = df[["Name", "Age"]].copy()

# 5-c  slicing
older_passengers = df[df["Age"] > 50]

display("Series head", names_series.head())
display("Mini-DF head", names_ages_df.head())
display("Over-50 passengers count:", older_passengers.shape[0])
```

```python
# 6.  Multi-index survival rate
# ==========================================
survival_table = (
    df.groupby(["Sex", "Pclass"])["Survived"]
      .agg(["mean", "count"])
      .round(3)
      .rename(columns={"mean": "SurvivalRate"})
)
survival_table
```

```python
# 7.  EDA plots
# ===========================================

# 7-a  Age histogram
plt.figure()
sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")
plt.title("Histogram of Passenger Ages")
plt.xlabel("Age"); plt.ylabel("Count")
plt.show()

# 7-b  Survival rate by Pclass
plt.figure()
sns.barplot(x="Pclass", y="Survived", data=df, palette="muted")
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# 7-c  Correlation heat-map
num_cols = df.select_dtypes(include=np.number).columns
plt.figure(figsize=(9, 7))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

# 7-d  Pair-plot (subset to avoid over-crowding)
cols = ["Age", "Fare", "SibSp", "Parch"]
sns.pairplot(df[cols + ["Survived"]], hue="Survived", diag_kind="kde")
plt.suptitle("Pair-Plot of Selected Features", y=1.02)
plt.show()
```

```python
# 8.  Survival rates by Sex & Class
# ==========================================
survival_by_sex = df.groupby("Sex")["Survived"].mean()
survival_by_class = df.groupby("Pclass")["Survived"].mean()

print("Survival by Sex:\n", survival_by_sex.round(3))
print("\nSurvival by Class:\n", survival_by_class.round(3))
```

```python
# 9.  Age distribution (box-plot)
# ==========================================
plt.figure()
sns.boxplot(x="Pclass", y="Age", hue="Sex", data=df, palette="Set2")
plt.title("Age Distribution by Passenger Class and Sex")
plt.show()
```

```python
# 10.  Fare distribution (violin-plot)
# ==========================================
plt.figure()
sns.violinplot(x="Pclass", y="Fare", hue="Survived", data=df, split=True, palette="pastel")
plt.title("Fare Distribution by Class and Survival Status")
plt.yscale("symlog")   # handle extreme fares
plt.show()
```
